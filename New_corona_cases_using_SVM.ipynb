```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('corona.csv')
data['Date']= pd.to_datetime(data['Date']) 
```

```python
data['month'] = data['Date'].dt.month
data['day'] = data['Date'].dt.day
data['year'] = data['Date'].dt.year
data['date_com'] = data['day'] + data['month'] 
```

```python
data
```

```python
data = data.drop(columns=['Date'])
data = data.drop(columns=['date_com'])
data = data.drop(columns=['year'])
```

```python
data
```

```python
#Finding the number of new cases in a particular date
#X = month and date
#Y = New_cases
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(data['month'], data[' New_cases'])
plt.show()
```

```python
X = data.iloc[:,[4,5]].values
Y = data.iloc[:,[0]].values
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Y = sc.fit_transform(Y)
Y
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'poly')
regressor.fit(X,Y)
```

```python
res1 = sc.inverse_transform(regressor.predict([[10,2]]))
```
